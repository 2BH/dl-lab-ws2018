```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import json
import seaborn
```

```python
# load the results from files
results = []
for i in range(4):
    with open('results/results_run_' + str(i) + '.json') as json_data:
        d = json.load(json_data)
        results.append(d)
        json_data.close()
```

```python
results[0]#["lr"]
```

```python
for i in range(4):
    learning_curve = results[i]["learning_curve"]
    plt.plot(np.arange(len(learning_curve))+1, learning_curve, label = "lr:%f" %results[i]["lr"])
plt.legend()
plt.xlabel("# epoches")
plt.ylabel("accuracy")
plt.grid(True)
plt.show()
```
